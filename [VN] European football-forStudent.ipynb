## Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp  theo số lượng giảm dần

- Tổng số bàn thắng của một đội bao gồm cả số bàn thắng ở sân nhà và sân khách.
- Sử dụng câu lệnh Union  

```python
home = pd.read_sql("""select t.team_long_name team_long_name, sum(m.home_team_goal) as Sum_Goal 
                            from team t 
                            join match m 
                                on m.home_team_api_id = t.team_api_id 
                            group by t.team_long_name
                            """,conn)
home
```

```python
team_goal = pd.read_sql("""with CTE as 
                            (select t.team_long_name team_long_name, sum(m.home_team_goal) as Sum_Goal 
                            from team t 
                            join match m 
                                on m.home_team_api_id = t.team_api_id 
                            group by t.team_long_name 
                            union all
                            select t.team_long_name team_long_name , sum(m.away_team_goal) as Sum_Goal 
                            from team t 
                            join match m 
                                on m.away_team_api_id = t.team_api_id 
                            group by t.team_long_name)
                            select team_long_name, sum(Sum_Goal) summ from cte
                            group by team_long_name
                            order by summ desc
                            """, conn)
team_goal
```

## Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.

- Nếu một đội có số lượng bàn thắng trên sân nhà > số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thắng" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà < số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thua" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà = số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "hòa" trong trận đấu này
- Sử dụng câu lệnh "Case When"


```python
# ĐIỀN CODE VÀO ĐÂY
sample = pd.read_sql("""select t.team_long_name ,
                        count(case when ((t.team_api_id = m.home_team_api_id and m.home_team_goal > m.away_team_goal) 
                            or (t.team_api_id = m.away_team_api_id and m.home_team_goal < m.away_team_goal)) then 1 end) as count_win,
                        COUNT(case when((team_api_id = m.home_team_api_id and m.home_team_goal < m.away_team_goal) 
                            or (t.team_api_id = m.away_team_api_id and m.home_team_goal > m.away_team_goal)) then 1 end) as count_lose,
                        COUNT(case when (team_api_id in (m.home_team_api_id, m.away_team_api_id) 
                            and m.home_team_goal = m.away_team_goal) then 1 end) as count_draw
                        from team t join match m on t.team_api_id = m.home_team_api_id or t.team_api_id = m.away_team_api_id
                        group by t.team_long_name
                    """,conn)
sample
```

## Câu hỏi 16: Liệt kê tỉ lệ phần trăm về số trận của mỗi giải trên tổng tất cả các trận

```python
# ĐIỀN CODE VÀO ĐÂY
pd.read_sql(""" with cte as(
            Select m.league_id,l.name, count(*) as 'COUNT' 
            from match m join league l 
            on m.league_id = l.id group by m.league_id, l.name)
            select name, round(100.0*count/(select sum(count) from cte),2) as 'percentage %'
            from cte
            group by name
            
            """,conn)
```

## Câu hỏi 17: Liệt kê tỉ lệ phần trăm về số bàn ghi được của từng giải so với tổng số bàn của tất cả các trận.

```python
# ĐIỀN CODE VÀO ĐÂY
pd.read_sql("""with cte as (select l.name, sum(m.home_team_goal + m.away_team_goal) as sum_goal
            from match m join league l on l.id = m.league_id
            group by l.name)
            select name, round(100.0*sum_goal/(select sum(sum_goal) from cte),2) as 'percentage %'
            from cte
            group by name
            
            """,conn)
```
